```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
kinney_data = pd.read_csv('rnap_tau_final_all_41.txt',sep ='\t', usecols = range(1,5))
kinney_data
```

By the code above, we can easily get any tab or comma separated data as rows and columns.

```python
#Lets say A = 0 , C = 1, G = 2, T = 3:
Random_Gene_Database = {}
Gene_Variable_List = ["A", "C", "G", "T"]
Tested_Gene = "" 
Binding_Energy = 0
for j in range(0, 100):
    for i in range(0, len(kinney_data)):
        Binding_Gene_Variable = np.random.randint(0,4) #writing inside the for loop, 
        #therefore it is randomly generated on every iteration. but this could be statistically biased.
        print("Row number:", i)
        print("The number is: ", Binding_Gene_Variable, " therefore, this is", Gene_Variable_List[Binding_Gene_Variable], ".")
        print("Required energy is: ", kinney_data.iloc[i, Binding_Gene_Variable], "joules. ")
        Binding_Energy += kinney_data.iloc[i, Binding_Gene_Variable]
        Tested_Gene += Gene_Variable_List[Binding_Gene_Variable]

    print("Our Gene Was: ", Tested_Gene)
    print("Total Energy: ", round(Binding_Energy, 4))

    Random_Gene_Database[Tested_Gene] = round(Binding_Energy, 4)
    Tested_Gene = ""
    Binding_Energy = 0
```

The code above gives us a random or many random 40 line Gene data and saves it in a dictionary with it's total binding energy. 
Data doesn't use the binding probabilities hence, it is not totally accurate.
